## Qs.1

```python
df.target.value_counts()
```

```python
df.target.value_counts().plot(kind='bar',color=['orchid','salmon'],width=0.1)
plt.title('Heart disease values')
plt.xlabel('1=Disease,0=No disease')
plt.ylabel('Amount');
```

### Q.3

```python
df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind='bar',color=['coral','lightskyblue','plum','khaki'])
plt.title('Chest pain vs no. of people');
```

```python
pd.crosstab(df.sex,df.cp)
```

```python
pd.crosstab(df.sex,df.cp).plot(kind='bar',color=['coral','lightskyblue','plum','khaki'])
plt.title('Types of chest pain w.r.x. sex')
plt.xlabel('0 = Female, 1=Male');
```

### 5 & 6

```python
sns.displot(x='age',data=df,bins=30,kde=True)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df.age[df.target==1],
           df.thalach[df.target==1],
           c='tomato')
plt.scatter(df.age[df.target==0],
           df.thalach[df.target==0],
           c='lightgreen')
plt.title('Max heart rate vs age')
plt.xlabel('age')
plt.ylabel('max heart rate')
plt.legend(['Disease','No disease']);
```

### Q.7&8

```python
plt.figure(figsize=(10,6))
plt.scatter(df.age[df.target==1],
           df.chol[df.target==1],
           c='salmon')
plt.scatter(df.age[df.target==0],
           df.chol[df.target==0],
           c='lightblue')
plt.title('Serum cholestroral vs age')
plt.xlabel('age')
plt.ylabel('Serum cholestroral')
plt.legend(['Disease','No disease']);
```
